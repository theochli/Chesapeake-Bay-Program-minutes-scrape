# For Loop checking each meeting for if the sponsor was the WQGIT

```python
all_mtgs_df.head(20)
```

```python
all_mtgs_df.shape
```

```python
sponsors = []

for i, row in all_mtgs_df.iterrows():
    cur_URL = row['mtg_details_url']
    
    page = requests.get(cur_URL)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="main-content")
    
    files = results.find_all('li') 
    
    if len(files) > 0:
        
        # check if last hyperlink is a file (has a file extension) or a link (likely a sponsor page)
        # if it has a file extension, then spliting on a '.' will result in a list of len 2
        # if a hyperlink to a sponsor page, then splitting on a '.' results in a list of len 1
        
        if len(files[-1].find('a')['href'].split('/')[-1].split('.')) == 1:           
            #append the name of the sponsor
            sponsors.append(files[-1].text)
        else:
            sponsors.append(None)
    
    else:
        sponsors.append(None)
        
```

```python
all_mtgs_df['sponsor'] = sponsors
```
